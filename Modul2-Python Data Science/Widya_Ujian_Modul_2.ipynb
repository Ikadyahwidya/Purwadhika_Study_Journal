If you realize, some columns that are numerical is listed as object data type. This is because in the data there is '?' value that become the placeholder of Null value. Now, replace all the '?' in every columns as NaN value (np.nan).

```python
for i in dfauto.columns:
    dfauto[i].replace('?',np.nan,inplace=True)
```

Now we will try to fill the NaN value for some of the columns.
For the NaN in Num-of-Doors, Bore, and  Stroke columns fill the NaN value with the mean of their columns with the similar value in the Make and Body-Style columns. *e.g if you have NaN in num-of-doors with hatchback renault car then fill the NaN with the num-of-doors mean of the hatchback renault car. 
 

```python
groupby_mean = dfauto.groupby(['Make','Body-Style']).mean()
```

```python
for col in ['Num-of-Doors','Bore','Stroke'] :
    dfauto[col] = dfauto[[col,'Make','Body-Style']].apply(lambda x : groupby_mean.loc[(x['Make'],
                                                                                       x['Body-Style']),col]
                                                          if pd.isna(x[col]) else x[col],axis = 1)
```

## Question 1

Do you think there is a relation between the price column and the Make column? Make you reasoning

## Answer 1 :
Hubungannya adalah :
Pembuatan (Make) mobil berdasarkan Supply in Demand dan berdasarkan apakah mobil itu termasuk golongan 'Kelas Atas' atau golongan 'Kelas Menengah ke bawah'.

Jika permintaan banyak, maka harga mobil yang dibuat semakin murah,
jika mobil itu termasuk golongan 'Kelas Atas' seperti mobil-mobil Eropa seperti BMW, Mercedes-benz, and Audi maka harga semakin mahal karena permintaan akan barang tersebut sedikit.
Hal tersebut dapat kita lihat dari gambar plot distrbution diatas : Semakin banyak merk mobil yang dibuat, maka harga mobil dengan merk tersebut semakin rendah, sebaliknya semakin sedikit merk mobil yang dibuat, maka harga mobil dengan merk tersebut semakin tinggi.


Create a new column called MPG-difference where the value is coming from the difference between Highway-MPG and City-MPG, then create another column called Car-Area from multipication of Length, Width, and Height columns.

```python
dfauto['MPG-difference']=abs(dfauto['Highway-MPG']-dfauto['City-MPG'])
dfauto['Car-Area']=dfauto['Length']*dfauto['Width']*dfauto['Height']
dfauto
```

```python
sns.distplot(dfauto['MPG-difference'])
```

```python
sns.distplot(dfauto['Car-Area'])
```

Now, use the IQR method to find the row outlier based on the MPG-Difference columns

```python
Q1=dfauto['MPG-difference'].describe()['25%']
Q3=dfauto['MPG-difference'].describe()['75%']
IQR=Q3-Q1
upper_limit=Q3+1.5*IQR
lower_limit=Q1-1.5*IQR
```
