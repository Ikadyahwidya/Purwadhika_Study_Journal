## Unsupervised Learning

We have lot of feature to be put into our unsupervised model, because of that let's use PCA to reduce our feature to 2 feature with random_state = 101. PC1 and PC2. 

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 2, random_state = 101)
```

```python
x_pca = pca.fit_transform(X)
```

```python
df_pca = pd.DataFrame(x_pca, columns = ['PC1', 'PC2'])
df_pca
```

```python
pca.explained_variance_ratio_
```

Our 2 PC only explained around 23 percent variance of the data, but we could just work with this data right now. Let's use the scatter plot to see the distribution of the 2 PC with the class feature as the hue

```python
plt.figure(figsize = (8,8))
sns.scatterplot('PC1','PC2',dfcancer['class'],data=pd.concat([df_pca,y],axis = 1))
```

Seems our 2 PC resulting in some scattered group, our class feature is not that separated but we could use these PCA result for our next analysis. 

Now, we would use KMeans, DBSCAN, and HDBSCAN to cluster our 2 PC. This time, I already done some analysis and find the best parameter we could use. Try to use the KMeans model with n_cluster = 5, DBSCAN model with eps = 0.1 and min_samples = 2, and HDBSCAN model with min_cluster_size =5 and min_samples = 2

Then create the scatter plot to compare the separation performance of each model. 

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 5)
kmeans.fit(df_pca[['PC1', 'PC2']])
df_pca['KM'] = kmeans.labels_
```

```python
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps = 0.1, min_samples = 2)
dbscan.fit(df_pca[['PC1', 'PC2']])
df_pca['DB'] = dbscan.labels_
```

```python
from hdbscan import HDBSCAN
hdbscan = HDBSCAN(min_cluster_size = 5, min_samples = 2)
hdbscan.fit(df_pca[['PC1', 'PC2']])
df_pca['HDB'] = hdbscan.labels_
```

```python
f, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4, sharey = True, figsize = (16, 8))

# Original scatter plot
ax1.set_title('Original')
ax1.scatter(df_pca['PC1'], df_pca['PC2'])

centroids = kmeans.cluster_centers_
# K Means scatter plot
ax2.set_title('K Means')
ax2.scatter(df_pca['PC1'], df_pca['PC2'], c = df_pca['KM'], cmap = 'rainbow')
ax2.scatter(centroids[:, 0], centroids[:, 1], marker = '*', s = 250, color = 'blue')

# DBSCAN scatter plot
core = dbscan.components_
ax3.set_title('DBSCAN')
ax3.scatter(df_pca['PC1'], df_pca['PC2'], c = df_pca['DB'], cmap = 'rainbow')
ax3.scatter(core[:, 0], core[:, 1], marker = 'o', c = 'black', s = 5)

# HDBSCAN scatter plot
ax4.set_title('HDBSCAN')
ax4.scatter(df_pca['PC1'], df_pca['PC2'], c = df_pca['HDB'], cmap = 'rainbow')
```

```python
dfcancer_kmeans = pd.concat([dfcancer, df_pca['KM']], axis = 1)
dfcancer_kmeans.rename(columns = {'KM' : 'k-label'}, inplace = True)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score

def calc_acc(X_train, y_train, X_test, y_test, model):
    model.fit(X_train, y_train)
    return accuracy_score(y_train, model.predict(X_train)), accuracy_score(y_test, model.predict(X_test))
```

### Logistic Regression

```python
logistic = LogisticRegression(random_state = 101)
print(f'Accuracy Train Original Data: {calc_acc(X_train, y_train, X_test, y_test, logistic)[0]}')
print(f'Accuracy Test Original Data: {calc_acc(X_train, y_train, X_test, y_test, logistic)[1]}')
print(f'Accuracy Train Oversampled Data: {calc_acc(X_train_os, y_train_os, X_test, y_test, logistic)[0]}')
print(f'Accuracy Test Oversampled Data: {calc_acc(X_train_os, y_train_os, X_test, y_test, logistic)[1]}')
```

### Random Forest

```python
rfc = RandomForestClassifier(random_state = 101)
print(f'Accuracy Train Original Data: {calc_acc(X_train, y_train, X_test, y_test, rfc)[0]}')
print(f'Accuracy Test Original Data: {calc_acc(X_train, y_train, X_test, y_test, rfc)[1]}')
print(f'Accuracy Train Oversampled Data: {calc_acc(X_train_os, y_train_os, X_test, y_test, rfc)[0]}')
print(f'Accuracy Test Oversampled Data: {calc_acc(X_train_os, y_train_os, X_test, y_test, rfc)[1]}')
```

### Decision Tree

```python
dtree = DecisionTreeClassifier(random_state = 101)
print(f'Accuracy Train Original Data: {calc_acc(X_train, y_train, X_test, y_test, dtree)[0]}')
print(f'Accuracy Test Original Data: {calc_acc(X_train, y_train, X_test, y_test, dtree)[1]}')
print(f'Accuracy Train Oversampled Data: {calc_acc(X_train_os, y_train_os, X_test, y_test, dtree)[0]}')
print(f'Accuracy Test Oversampled Data: {calc_acc(X_train_os, y_train_os, X_test, y_test, dtree)[1]}')
```

### XGBoost

```python
xgb = XGBClassifier(random_state = 101)
print(f'Accuracy Train Original Data: {calc_acc(X_train, y_train, X_test, y_test, xgb)[0]}')
print(f'Accuracy Test Original Data: {calc_acc(X_train, y_train, X_test, y_test, xgb)[1]}')
print(f'Accuracy Train Oversampled Data: {calc_acc(X_train_os, y_train_os, X_test, y_test, xgb)[0]}')
print(f'Accuracy Test Oversampled Data: {calc_acc(X_train_os, y_train_os, X_test, y_test, xgb)[1]}')
```

### Cross Validation Using Original Data

```python
from sklearn.model_selection import KFold
from sklearn.metrics import matthews_corrcoef, f1_score, classification_report, roc_auc_score, confusion_matrix, log_loss
kf = KFold(n_splits = 5, shuffle = True, random_state = 42)
```

```python
def calc_train_error(X_train, y_train, model):
    predictions = model.predict(X_train)
    predictProba = model.predict_proba(X_train)
    matt = matthews_corrcoef(y_train, predictions)
    f1 = f1_score(y_train, predictions, average = 'macro')
    report = classification_report(y_train, predictions)
    roc_auc = roc_auc_score(y_train, predictProba[:, 1])
    accuracy = accuracy_score(y_train, predictions)
    confMatrix = confusion_matrix(y_train, predictions)
    logloss = log_loss(y_train, predictProba)
    return{
        'report' : report,
        'matthew' : matt,
        'f1' : f1,
        'roc' : roc_auc,
        'accuracy' : accuracy,
        'confusion' : confMatrix,
        'logloss' : logloss
    }

def calc_validation_error(X_test, y_test, model):
    predictions = model.predict(X_test)
    predictProba = model.predict_proba(X_test)
    matt = matthews_corrcoef(y_test, predictions)
    f1 = f1_score(y_test, predictions, average = 'macro')
    report = classification_report(y_test, predictions)
    roc_auc = roc_auc_score(y_test, predictProba[:, 1])
    accuracy = accuracy_score(y_test, predictions)
    confMatrix = confusion_matrix(y_test, predictions)
    logloss = log_loss(y_test, predictProba)
    return{
        'report' : report,
        'matthew' : matt,
        'f1' : f1,
        'roc' : roc_auc,
        'accuracy' : accuracy,
        'confusion' : confMatrix,
        'logloss' : logloss
    }

def calc_metrics(X_train, y_train, X_test, y_test, model):
    model.fit(X_train, y_train)
    train_error = calc_train_error(X_train, y_train, model)
    validation_error = calc_validation_error(X_test, y_test, model)
    return train_error, validation_error
```

```python
train_errors = []
validation_errors = []
for train_index, val_index in kf.split(X,y):
    # split data
    X_train, X_val = X.iloc[train_index], X.iloc[val_index]
    y_train, y_val = y.iloc[train_index], y.iloc[val_index]
    # instantiative model
    rfc_new = RandomForestClassifier(n_estimators = 100, min_samples_split = 20, min_samples_leaf = 1,
                                max_features = 'sqrt', max_depth = 10, criterion = 'entropy', bootstrap = False)
    # calculate errors
    train_error, val_error = calc_metrics(X_train, y_train, X_val, y_val, rfc_new)
    # append to appropriate list
    train_errors.append(train_error)
    validation_errors.append(val_error)
```

```python
for i, tr_err in zip(range(1, 6), validation_errors):
    print(f'Report Train ke {i}')
    print(tr_err['report'])
```

```python
for i, val_err in zip(range(1, 6), validation_errors):
    print(f'Report Validation ke {i}')
    print(val_err['report'])
```

```python
data = list()
for tr_err, val_err in zip(train_errors, validation_errors):
    data_row = list()
    for param in 'accuracy roc f1 matthew logloss'.split():
        data_row.append('')
    data.append([tr_err['accuracy'], val_err['accuracy'],
               tr_err['roc'], val_err['roc'],
               tr_err['f1'], val_err['f1'],
               tr_err['matthew'], val_err['matthew'],
               tr_err['logloss'], val_err['logloss']])

tabel_1 = pd.DataFrame(data, index = ['1 Iteration', '2 Iteration', '3 Iteration', '4 Iteration', '5 Iteration'],
                    columns = ['Train Accuracy', 'Test Accuracy', 'Train ROC AUC', 'Test ROC AUC',
                              'Train F1 Score', 'Test F1 Score', 'Train Matthews Corr Coef', 'Test Matthews Corr Coef',
                              'Train Log Loss', 'Test Log Loss'])

mean = list()
for i in tabel_1:
    mean.append(tabel_1[i].mean())
    
tabel_2 = pd.DataFrame(mean, index = tabel_1.columns, columns = ['Average'])

pd.concat([tabel_1, tabel_2.transpose()]) 
```

Melihat ringkasan di atas, saya menyimpulkan bahwa Model Random Forest cukup stabil untuk memprediksi accuracy score, F1 Score karena di setiap iterasi model dapat menghasilkan nilai yang konsisten pada kedua metric tersebut jika dibandingkan dengan hasil dari metric lainnya pada data test. Jika kita lihat hasil pada data test menggunakan ROC_AUC terjadi overfitting yang sangat buruk, yaitu pada iterasi 1 dan iterasi 2, yaitu dari 0.8333 turun menjadi 0.555.

### Learning Curve

```python
from sklearn.model_selection import learning_curve

train_sizes, train_scores, test_scores = learning_curve(estimator=rfc_new,
                                                       X=X,
                                                       y=y,
                                                       train_sizes=np.linspace(0.5, 1.0, 6),
                                                       cv=10)

print(train_scores)
# Mean value of accuracy against training data
train_mean = np.mean(train_scores, axis=1)
print(train_mean)
print(train_sizes)
# Standard deviation of training accuracy per number of training samples
train_std = np.std(train_scores, axis=1)

# Same as above for test data
test_mean = np.mean(test_scores, axis=1)
test_std = np.std(test_scores, axis=1)

# Plot training accuracies 
plt.plot(train_sizes, train_mean, color='red', marker='o', label='Training Accuracy')
# Plot the variance of training accuracies
plt.fill_between(train_sizes,
                train_mean + train_std,
                train_mean - train_std,
                alpha=0.15, color='red')

# Plot for test data as training data
plt.plot(train_sizes, test_mean, color='blue', linestyle='--', marker='s', 
        label='Test Accuracy')
plt.fill_between(train_sizes,
                test_mean + test_std,
                test_mean - test_std,
                alpha=0.15, color='blue')

plt.xlabel('Number of training samples')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

Melihat plot di atas, sepertinya model ini cukup baik.
